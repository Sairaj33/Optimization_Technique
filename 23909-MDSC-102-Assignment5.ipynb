BOXPLOT

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv(r"C:\Users\Msc 1\Desktop\wine\wine.data")
dataset
```

```python
dataset.columns =['Class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols',
'Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline' ]
dataset
```

```python
sns.boxplot(x=dataset["Ash"])
```

```python
i=1
plt.figure(figsize=(20,20))
for feature in dataset:
    if feature!='Class' and feature!='Alcohol':
        plt.subplot(4,3,i)
        sns.boxplot(x=dataset['Class'],y=dataset[feature], hue=dataset['Class'])
        i+=1
```

```python
i=1
plt.figure(figsize=(20,20))
for feature in dataset:
    if feature!='Class' and feature!='Alcohol':
        plt.subplot(4,3,i)
        sns.histplot(x=dataset[feature], hue=dataset['Class'])
        i+=1
```
