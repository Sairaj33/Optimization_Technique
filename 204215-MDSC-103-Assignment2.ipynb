```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-5,5,100)
f = lambda x: x**2

plt.plot(x,f(x))

eta=0.1
x= 4
err = 1e-5
i=0
arr = []
while i<100:
    x1 = x-(eta*(2*x))
    if(np.abs(x1-x)<err):
        break
    x = x1
    arr.append(x)
    i+=1 
    #print(x)
    plt.scatter(x,f(x))
print(x)
plt.plot(x,f(x), color='r')                         # plot for x^2
plt.plot( color='b',marker ='o') 
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.title("Gardient Descent for X^2")
plt.legend(["y = x^2","Gradient descent"])
```

```python
def gd_func (fun,df,int_pt,lr=0.01,err=1e-5):
    i=0
    x1=0
    arr=[]

    while(i<100):

        x1= int_pt - lr*df(int_pt)
        
        if((int_pt - x1) < err):
            print('point', x1)
            break 
        else:
            int_pt = x1
            i+=1
        arr.append(int_pt)

    X_val = np.arange(-5,5,.1)
    e = np.array([fun(x) for x in X_val])

    
    plt.plot(X_val,e)

    
    plt.xlabel("X values")
    plt.ylabel("f(x)")
    plt.title("f(x) = x^3")

    for x in arr:
        plt.scatter(x,fun(x))
    
    
```

```python
fun = lambda x: x**3
df = lambda x: 3* x**2
```

```python
gd_func(fun,df,5)
```
