```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
import math
```

*Binomial Distribution*

```python
n, p = 30, 0.2
x = np.arange(0, n + 1)
binomial_dist = np.array([np.math.comb(n, k) * (p ** k) * ((1 - p) ** (n - k)) for k in x])

plt.bar(x, binomial_dist)
plt.plot(x, binomial_dist)
plt.xlabel("Number of Successes")
plt.ylabel("Probability")
plt.title(f"Binomial Distribution (n={n}, p={p})")

```

Poission Distribution

```python
def poisson_pmf(k, lambda_param):
    return (np.exp(-lambda_param) * (lambda_param ** k)) / np.math.factorial(k)

def plot_poisson_distribution(lambda_param, max_value):
    # Generate the Poisson distribution
    x = np.arange(0, max_value + 1)
    poisson_dist = np.array([poisson_pmf(k, lambda_param) for k in x])

    plt.bar(x, poisson_dist)
    plt.xlabel("Number of Events")
    plt.ylabel("Probability")
    plt.title(f"Poisson Distribution (lambda={lambda_param})")
    plt.show()
#lambda_param = 0.01
lambda_param = 5
#lambda_param = 1
max_value_to_plot = 20 
plot_poisson_distribution(lambda_param, max_value_to_plot)
```

Exponential Distribution

```python
lambda_param = 0.9
x = np.linspace(0, 10, 1000)
exponential_dist = lambda_param * np.exp(-lambda_param * x)

plt.bar(x, exponential_dist)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title(f"Exponential Distribution (lambda={lambda_param})")
plt.show()
```

Normal Distributions

```python
m = 1 #mean
s = 3 #std_dev
x = np.linspace(-5, 5, 1000)
normal_distribution = (1 / (s * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - m) / s) ** 2)

plt.bar(x, normal_distribution)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (mean={m}, std_dev={s})")
plt.show()
```

Gamma Distribution

```python

alpha = 2
lambda_param = 1
x = np.linspace(0, 10, 1000)
gamma_dist = (x ** (alpha - 1)) * np.exp(-x / lambda_param) / (lambda_param ** alpha * np.math.gamma(lambda_param))

plt.bar(x, gamma_dist)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title(f"Gamma Distribution (shape={alpha}, scale={lambda_param})")
plt.show()
```
