```python
dataset.columns =['Class','Alcohol','Malic acid','Ash','Alcalinity of ash','Magnesium','Total phenols','Flavanoids','Nonflavanoid phenols',
'Proanthocyanins','Color intensity','Hue','OD280/OD315 of diluted wines','Proline' ]
dataset
```

```python
import seaborn as sns
```

```python
sns.countplot(x=dataset.Class)
```

```python
fig,axes=plt.subplots(4,3,figsize=(20,20))

sns.scatterplot(x=dataset['Alcohol'], y= dataset['Malic acid'], hue=dataset.Class,ax=axes[0,0])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Ash'], hue=dataset.Class,ax=axes[0,1])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Alcalinity of ash'], hue=dataset.Class,ax=axes[0,2])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Magnesium'], hue=dataset.Class,ax=axes[1,0])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Total phenols'], hue=dataset.Class,ax=axes[1,1])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Flavanoids'], hue=dataset.Class,ax=axes[1,2])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Nonflavanoid phenols'], hue=dataset.Class,ax=axes[2,0])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Proanthocyanins'], hue=dataset.Class,ax=axes[2,1])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Color intensity'], hue=dataset.Class,ax=axes[2,2])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Hue'], hue=dataset.Class,ax=axes[3,0])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['OD280/OD315 of diluted wines'], hue=dataset.Class,ax=axes[3,1])
sns.scatterplot(x=dataset['Alcohol'], y= dataset['Proline'], hue=dataset.Class,ax=axes[3,2])
```
